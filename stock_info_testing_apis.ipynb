### This Function will give all the stocks for an exchange. We might use this one to check if we can analyse a stock

```python
def get_all_stocks_by_exchange(exchange):
	end_url_path="stocks"
	url = f"https://twelve-data1.p.rapidapi.com/{end_url_path}"

	querystring = {"exchange":exchange,"format":"json"}

	headers = {
		"X-RapidAPI-Key": "628945fc0cmsh40c2bcb4a06a38ap16df3cjsn9834d9b0ab25",
		"X-RapidAPI-Host": "twelve-data1.p.rapidapi.com"
	}
	response = requests.get(url, headers=headers, params=querystring)
	data = response.json()
	df = pd.DataFrame(data['data'])
	print(df.count())
	print(df.columns)
	return df
```

```python
df = get_all_stocks_by_exchange('NASDAQ')
df[df.name.str.lower().str.contains('nano')]
```
